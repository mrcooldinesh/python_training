```
# Given a date, identify number of days past the given date

from matplotlib.axis import datetime

day1 = datetime.date(1987, 8, 23)
day2 = datetime.date.today()
daysBetween = (day2 - day1)

print(f'Number of Days between {day1} and today {day2} : {daysBetween.days}')
print(f'Number of Seconds between {day1} and today {day2} : {daysBetween.total_seconds()}')
```

```
'''
Creat list of numbers divisible by 5 from 10 to 100
and then function to find sum of values divisible by 10
'''

# Function to print sum of numbers divisible by 10 give list as input
def sumofdiv10(a):
  t = 0
  for i in a: #Block1
    if(i%10 == 0): #Block2
      t = t+i
  
  print(f'Total sum of values divisible by 10 : {t}')

l1 = [x for x in range(10,101)]  #Generating list of numbers from 10 to 100
print(f'{l1}\n')

newlist = []  #Creating Empty list

for i in l1:  #Block1
  if(i%5 == 0):  #Block2
    newlist.append(i) #Adding each item to the list
print('###### List of Values Divisible by 5 ######\n')
print(newlist)

print('\n###### Sum of Values Divisible by 10 ######\n')
s = sumofdiv10(l1)  #Calling function
```

```
def do_normalise(crop_img):
  return -np.log(1/((1+ crop_img)/257) -1)

def undo_normalise(crop_img):
  return (1+ 1/(np.exp(-crop_img)+ 1)* 257).astype("uint8")

def rotation_matrix(theta):
  return np.c_[
      [1,0,0],
      [0,np.cos(theta),-np.sin(theta)],
      [0,np.sin(theta),np.cos(theta)]
  ]

img_norm = do_normalise(crop_img)
img_rot = np.einsum("ijk,lk->ijl", img_norm, rotation_matrix(np.pi))
img = undo_normalise(img_rot)

imgplot = plt.imshow(img)
```
